```python
# import the pandas library 
import pandas as pd 
# read the file into a pandas dataframe 
df = pd.read_csv('cneos_fireball_data.csv')
# drop vx, vy, and vz columns
df = df.drop(columns = ['vx', 'vy', 'vz'])
# drop rows that have any missing values 
df = df.dropna()
df
```

```python
def date():
# Create dictionary with date as key and all other categories as values
    date = df.set_index('Peak Brightness Date/Time (UT)').T.to_dict('list')
# Ask user to enter a date
    date_input = str(input("Please enter a date in the format 'year-month-day hour:min:sec': "))
# Scan dictionary that prints all values at specified date
# Return dictionary
    print(date[date_input])
date()
```

```python
def altitude():
# Create dictionary with altitude as key and all other categories as values
    alt = df.set_index('Altitude (km)’).T.to_dict('list')
# Ask user to enter an altitude
    alt_input = str(input("Please enter an altitude value: "))
# Scan dictionary that prints all values at specified altitude
# Return dictionary
    print(alt[alt_input])
altitude()
```

```python
def velocity():
# Create dictionary with velocity as key and all other categories as values
    vel = df.set_index('Velocity (km/s)’).T.to_dict('list')
# Ask user to enter a velocity
    vel_input = str(input("Please enter a velocity value: "))
# Scan dictionary that prints all values at specified velocity
    print(vel[vel_input])
# Return dictionary
velocity()
```
